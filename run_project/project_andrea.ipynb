# Set

```python
from models.discriminator import Discriminator
import utils.getters_setters as gs
import torch
import argparse

args = argparse.Namespace()
args.model = 'deeplabv3_mobilenetv2'
args.dataset = 'gta'
args.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
args.path_model = '5a_model_smart-smoke-524_epoch5_0.263'
args.path_discriminator = '5a_discriminator_smart-smoke-524_epoch7_0.710'
args.bs =  1
args.hnm = True
args.step = '5a'
args.cjvalue = 0.3
args.lr = 0.01
args.m = 0.9
args.wd = 1e-10
args.lam_adv = 1e-6
args.conf_th = 0.8
args.fraction = 0
args.alpha_kd = 0.5

model = gs.model_init(args)
model.to(args.device)
discriminator = Discriminator(in_channels=gs.get_dataset_num_classes(args.dataset))
discriminator.to(args.device)

train_datasets, test_datasets, eval_datasets = gs.get_datasets(args)
train_clients, test_clients = gs.gen_clients(args, eval_datasets, test_datasets, model, discriminator)
metrics = gs.set_metrics(args)

gs.loadCheckpoint(path=args.path_model, model=model, device=args.device)
# gs.loadCheckpoint(path=args.path_discriminator, model=discriminator, device=args.device)
```

# Try adversarial client

```python
from utils.advTrainer import AdvTrainer
from advClient import AdvClient
from torch.nn import Softmax2d, LogSoftmax

self_trainer = AdvTrainer(discriminator, conf_th=args.conf_th, fraction=args.fraction)
self_trainer.set_teacher(model)

c = train_clients[0]

model.train()
discriminator.eval()

for batch_id, (images, labels) in enumerate(c.train_loader):

    images, labels = images.to(args.device), labels.to(args.device)
    
    c.optimizerG.zero_grad()

    outputs = c.get_outputs(images)
    cpmap = Softmax2d()(outputs)
    cpmaplsmax = LogSoftmax(dim=1)(outputs)

    # substitute actual labels with the pseudo ones
    labels = self_trainer.get_pseudo_lab(images)
    print(labels[labels != 255].size())

    # semi supervised loss
    LGsemi = c.loss_function(cpmaplsmax,labels)
    
    conff = LogSoftmax(dim=1)(c.discriminator(cpmap))

    N, _, H, W = cpmap.size()
    targetr = torch.ones((N,H,W), dtype=torch.long).to(args.device)
    # adversarial loss
    LGadv = c.loss_function(conff, targetr)

    LGseg = LGsemi + args.lam_adv *LGadv

    LGseg.backward()
    
    c.optimizerG.step()

    print(f"{LGsemi.item() = }, {args.lam_adv*LGadv.item() = }")
    if batch_id > 5:
        break
```

# Try adversarial centralized

```python
from torch.nn import Softmax2d, LogSoftmax
from torch.nn import NLLLoss
from advCentralized import AdvCentralized

clf = AdvCentralized(args, model, discriminator, train_datasets, metrics['train'])

for batch_id, (images, labels) in enumerate(clf.train_loader):

    images, labels = images.to(clf.args.device), labels.to(clf.args.device)
    
    ohlabels = labels.detach().clone()
    # one-hot encoded label
    ohlabels[ohlabels == 255] = 15
    # print(label.unique())
    ohlabels = torch.nn.functional.one_hot(ohlabels).permute((0, 3, 1, 2))

    with torch.no_grad():
        outputs = clf.get_outputs(images)
        cpmap = Softmax2d()(outputs)

    N, _, H, W = cpmap.size()

    # Generate the Real and Fake Labels
    targetf = torch.zeros((N,H,W), dtype=torch.long, requires_grad=False).to(clf.args.device)
    targetr = torch.ones((N,H,W), dtype=torch.long, requires_grad=False).to(clf.args.device)

    ####################

    clf.optimizerD.zero_grad()

    # Train on Real
    confr = LogSoftmax(dim=1)(clf.discriminator(ohlabels.float()))
    # compute D loss on real
    LDr = clf.loss_function(confr,targetr)
    # compute gradient
    LDr.backward()

    # Train on Fake
    conff = LogSoftmax(dim=1)(clf.discriminator(cpmap.data))
    # compute D loss on G output (fake)
    LDf = clf.loss_function(conff,targetf)
    # compute gradient
    LDf.backward()

    clf.optimizerD.step()

    ###################

    clf.optimizerG.zero_grad()

    outputs = clf.get_outputs(images)
    cpmap = Softmax2d()(outputs)
    cpmaplsmax = LogSoftmax(dim=1)(outputs)
    conff = LogSoftmax(dim=1)(clf.discriminator(cpmap))


    LGce = clf.loss_function(cpmaplsmax,labels)
    LGadv = clf.loss_function(conff,targetr)
    print(f"{LGce.item() = }, {LGadv.item() = }")
    LGseg = LGce + clf.args.lam_adv *LGadv

    LGseg.backward()
    
    clf.optimizerG.step()

    if batch_id > 5:
        break
```
